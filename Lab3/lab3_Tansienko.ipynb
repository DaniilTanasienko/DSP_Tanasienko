3. Імпортувати RGB-зображення Image з диску. Оцінити його розмір у пікселях N x M та об’єм пам’яті, який він займає на диску.

5. Зменшити початкове RGB-зображення Image у n разів, візуалізувати його, перевіривши результат масштабування, та експортувати його на диск. Чи змінився об’єм пам’яті, який він займає на диску? Обгрунтувати отриманий результат.

```python
import cv2 as cv
import os
import matplotlib.pyplot as plt
import math

image_path = 'src/photo.png'   
n = 2                          
image_rgb = cv.imread(image_path)

if image_rgb is None:
    print(f"Не удалось открыть изображение: {image_path}")
else:
    height_N, width_M, _ = image_rgb.shape
    file_size_original = os.path.getsize(image_path)

    print(f"Результати:")
    print(f"Оригінал (N x M): {height_N} x {width_M}")
    print(f"Об'єм (оригінал): {file_size_original} байт ({file_size_original/1024:.1f} KB)\n")

    new_width = math.floor(width_M / n)
    new_height = math.floor(height_N / n)
    new_dim = (new_width, new_height)
    image_resized = cv.resize(image_rgb, new_dim, interpolation=cv.INTER_AREA)
     
    print(f"Зменшено у {n} рази:")
    print(f"Новий розмір:{new_height} x {new_width}\n")
    
    resized_path = 'src/photo_resized.png'
    # Сохраняем с сжатием (для PNG уровень сжатия от 0 до 9)
    cv.imwrite(resized_path, image_resized, [cv.IMWRITE_PNG_COMPRESSION, 9])     
    
    file_size_resized = os.path.getsize(resized_path)
    print(f"Новий об'єм:{file_size_resized} байт ({file_size_resized/1024:.1f} KB)\n")
    
    print(f"Результат")
    if file_size_resized < file_size_original:
        diff = file_size_original - file_size_resized
        ratio = file_size_original / file_size_resized
        print(f"Так, об'єм пам'яті зменшився на {diff} байт (у {ratio:.1f} разів).")
        print("Пояснення: для n=2 кількість пікселів зменшилася у 4 рази.")
        print("Менша кількість пікселів означає менший обсяг даних для зберігання, тому розмір файлу зменшився.")
    else:
        print("Ні, обсяг пам’яті не зменшився або навіть зріс.")
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(image_rgb, cv.COLOR_BGR2RGB))
    plt.title(f'Оригінал\n{height_N}x{width_M}')
    plt.subplot(1, 2, 2)
    plt.imshow(cv.cvtColor(image_resized, cv.COLOR_BGR2RGB))
    plt.title(f'Зменшене\n{new_height}x{new_width}')     
    plt.show()
```

6. Написати функцію, яка буде виконувати пропорційне зменшення початкового зображення розміром N x M пікселей відповідно до вимоги, щоби зменешене зображення мало ширину K пікселей.

```python
import cv2 as cv
import matplotlib.pyplot as plt

def resize_proportional(image, target_width):
    (h, w) = image.shape[:2]
    ratio = target_width / float(w)
    target_height = int(h * ratio)
    new_dimensions = (target_width, target_height)
    resized_img = cv.resize(image, new_dimensions, interpolation=cv.INTER_AREA)
    return resized_img

image_path = 'src/photo.png'
K = 100

image_bgr = cv.imread(image_path)

if image_bgr is None:
    print(f"Помилка: Не вдалося знайти файл {image_path}")
else:
    image_resized_k = resize_proportional(image_bgr, K)
    
    print(f"Результати Завдання 6")
    print(f"ширина K: {K} пікселів")
    print(f"Оригінал (h x w): {image_bgr.shape[0]} x {image_bgr.shape[1]}")
    print(f"Результат (h x w): {image_resized_k.shape[0]} x {image_resized_k.shape[1]}")
    
    plt.figure(figsize=(6, 6))
    plt.imshow(cv.cvtColor(image_resized_k, cv.COLOR_BGR2RGB))
    plt.title(f'Пропорційно зменшено')
    plt.show()    
```

7. Зробити фото розвороту власної залікової книжки з перспективним викривленням на фоні письмового столу. Вирізати зону із заліковкою та виконати афінні та перспективні перетворення за трьома точками точками і візуалізовані. В кожному випадку результати мають бути записані на диск.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image_path = 'src/Zalikova.jpg' 
image = cv.imread(image_path)

if image is None:
    print(f"Помилка {image_path}")
else:
    rows, cols, ch = image.shape
    print(f"Розмір зображення: {cols}x{rows}")
    pts_input_affine = np.float32([
        [415, 340],
        [573, 342],
        [410, 540]
    ])
    pts_output_affine = np.float32([
         [150, 220],
        [450, 210],
        [140, 600]
    ])
    M_affine = cv.getAffineTransform(pts_input_affine, pts_output_affine)
    img_affine = cv.warpAffine(image, M_affine, (cols, rows))
    pts_input_persp = np.float32([           
        [700, 400],  
        [1200, 400],  
        [700, 1100], 
        [1200, 1100]  
    ])  

    width_new, height_new = 300, 400    
    pts_output_persp = np.float32([
        [0, 0],
        [width_new, 0],
        [0, height_new],
        [width_new, height_new]
    ])

    M_persp = cv.getPerspectiveTransform(pts_input_persp, pts_output_persp)
    img_perspective = cv.warpPerspective(image, M_persp, (width_new, height_new))

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 3, 1)
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title("Оригінал")
    plt.grid(True)    
    plt.subplot(1, 3, 2)
    plt.imshow(cv.cvtColor(img_affine, cv.COLOR_BGR2RGB))
    plt.title("Афінне")
    plt.subplot(1, 3, 3)
    plt.imshow(cv.cvtColor(img_perspective, cv.COLOR_BGR2RGB))
    plt.title("Перспектива")

    plt.show()
```

1. У чому полягає сенс афінних перетворень?

Суть афінних перетворень полягає у зміні геометрії зображення зі збереженням важливої властивості — паралельності прямих. Це означає, що будь-які лінії, які були паралельними на початковому зображенні, залишаються паралельними й після перетворення. Наприклад, квадрат може бути перетворений на паралелограм, але не на трапецію.

2. Чи є операції трансляції, масштабування і повороту афінними? Відповідь обґрунтувати.

Так, ці операції належать до афінних. Афінне перетворення описується як лінійне перетворення координат із додаванням вектора зсуву.

Трансляція (зсув) полягає в додаванні сталого вектора до координат усіх точок, при цьому паралельність ліній не порушується.

Масштабування здійснюється множенням координат на певний коефіцієнт, що також не впливає на паралельність.

Поворот обертає всі точки зображення навколо заданого центру на фіксований кут, зберігаючи паралельність прямих.

Оскільки в усіх цих випадках паралельні лінії залишаються паралельними, трансляція, масштабування і поворот є афінними перетвореннями.

3. Чи є перспективні перетворення частинним випадком афінних перетворень? Відповідь обґрунтувати.

Ні, перспективні перетворення не є частинним випадком афінних. Навпаки, афінні перетворення можна вважати спрощеним окремим випадком проєктивних (перспективних) перетворень.

Ключова відмінність між ними полягає у збереженні паралельності. Афінні перетворення завжди зберігають паралельність прямих, тоді як перспективні — ні. У перспективному зображенні лінії, які є паралельними в реальності, візуально сходяться в одній точці на горизонті.

Таким чином, перспективні перетворення утворюють більш загальний і складний клас, який дозволяє перетворювати прямокутник у довільний чотирикутник, що неможливо при афінних перетвореннях.
